# `filtering dataframe`

```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.loc[df['Type 1'] == 'Dark']
```

```python
df.loc[((df['Type 1'] == 'Dark') | (df['Type 2'] == 'Dark')) & (df['Total'] > 500)]
```

```python
df.loc[(df['Type 1'] == 'Dark')| (df['Defense'] >= 500)]
```

```python
df.loc[(df['Type 1'] == 'Dark')| (df['Defense'] <= 500)]
```
