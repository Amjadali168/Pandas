```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

# `pokemon which are types of water`

```python
df.loc[df['Type 1'] == 'Water']
```

```python
# pokemon which are water as well as dark
df.loc[(df['Type 1'] == 'Water') & (df['Type 2'] == 'Dark')]
```

```python
# water and dark pokemons having HP  greater than 60
df.loc[(df['Type 1'] == 'Water') & (df['Type 2'] == 'Dark') & (df['HP'] > 60)]
```

```python
# fire pokemon which are dark, ghost or ground
array =['Dark','Ghost','Ground']
df.loc[(df['Type 1'] == 'Fire') & (df['Type 2'].isin(array))]
```

```python
# fire, water or rain pokemons
array = ['Fire','Water','Rain']
df.loc[df['Type 1'].isin(array)]
```
